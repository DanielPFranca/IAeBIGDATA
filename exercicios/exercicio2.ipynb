```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

quantidade = np.array([100,200,300,400,500,600,700]).reshape(-1,1)
temperatura = np.array([30,35,40,45,60,100,150]).reshape(-1,1)

modelo = LinearRegression()
modelo.fit(quantidade, temperatura)


quantidade_teste = np.array([100,200,50,300,1000]).reshape(-1,1)
temperatura_prevista = modelo.predict(quantidade_teste)

```

```python
for i in range(len(quantidade_teste)):
    print(f'Quantidade em J: {quantidade_teste[i][0]}  - Temperatura prevista: {temperatura_prevista[i]}')

```

```python
plt.scatter(quantidade, temperatura, color='blue', label='Dados de treinamentos') # gráficos com pontos
plt.plot(quantidade_teste, temperatura_prevista, color='red', label='Dados de teste')
plt.xlabel('Quantidade')
plt.ylabel('Temperatura em C°')
plt.title('Relação entre Quantidade e Temperatura em C°')
plt.legend() # exibe a legenda do gráfico
plt.grid(True) # exibe o grid do gráfico
plt.show()
```
