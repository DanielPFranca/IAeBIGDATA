```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

area = np.array([50,100,120,300]).reshape(-1,1)
valor = np.array([180.000,300.000,375.000,600.000]).reshape(-1,1)

modelo = LinearRegression()
modelo.fit(area, valor)


area_teste = np.array([100,200,50,300,500]).reshape(-1,1)
valor_previsto = modelo.predict(area_teste)
```

```python
for i in range(len(area_teste)):
    print(f'Área em m²: {area_teste[i][0]}  - Valor $$ previsto: {valor_previsto[i]} ')
```

```python
plt.scatter(area, valor, color='blue', label='Dados de treinamentos') # gráficos com pontos
plt.plot(area_teste, valor_previsto, color='red', label='Dados de teste')
plt.xlabel('Área')
plt.ylabel('Valor')
plt.title('Relação entre Área e Valor')
plt.legend() # exibe a legenda do gráfico
plt.grid(True) # exibe o grid do gráfico
plt.show()
```
