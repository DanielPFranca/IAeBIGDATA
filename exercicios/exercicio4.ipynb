```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

renda = np.array([1200,5000,8000,10000,30000]).reshape(-1,1)
gastos = np.array([750,1500,4000,6000,18000]).reshape(-1,1)

modelo = LinearRegression()
modelo.fit(renda, gastos)


renda_teste = np.array([1000,8000,34000,50000,47000]).reshape(-1,1)
gastos_previsto = modelo.predict(renda_teste)
```

```python
for i in range(len(renda_teste)):
    print(f'Renda $: {renda_teste[i][0]}  - Gastos $ previsto: {gastos_previsto[i]} ')
```

```python
plt.scatter(renda, gastos, color='blue', label='Dados de treinamentos') # gráficos com pontos
plt.plot(renda_teste, gastos_previsto, color='red', label='Dados de teste')
plt.xlabel('Renda')
plt.ylabel('Gastos')
plt.title('Relação entre Renda e Gastos')
plt.legend() # exibe a legenda do gráfico
plt.grid(True) # exibe o grid do gráfico
plt.show()
```
