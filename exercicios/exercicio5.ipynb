```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

consumo = np.array([50,140,30,75,123]).reshape(-1,1)
energia = np.array([60,150,50,85,100]).reshape(-1,1)

modelo = LinearRegression()
modelo.fit(consumo, energia)


consumo_teste = np.array([50,20,70,30,200]).reshape(-1,1)
energia_prevista = modelo.predict(consumo_teste)
```

```python
for i in range(len(consumo_teste)):
    print(f'Consumo: {consumo_teste[i][0]} KW - Energia privista: {energia_prevista[i]}')
```

```python


plt.scatter(consumo, energia, color='blue', label='Dados de treinamentos') # gráficos com pontos
plt.plot(consumo_teste, energia_prevista, color='red', label='Dados de teste')
plt.xlabel('Consumo KW')
plt.ylabel('Energia')
plt.title('Relação entre Consumo e Energia')
plt.legend() # exibe a legenda do gráfico
plt.grid(True) # exibe o grid do gráfico
plt.show()
```
