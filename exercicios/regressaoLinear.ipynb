```python
import numpy as np
from sklearn.linear_model import LinearRegression

vazao = np.array([10,20,30,5,35,40]).reshape(-1,1)
nivel = np.array([50,60,70,45,80,85]).reshape(-1,1)

modelo = LinearRegression()
modelo.fit(vazao, nivel)

#vazao_teste = np.array([15,25,33]).reshape(-1,1)
#vazao_teste = np.array([5,2,7,30,50]).reshape(-1,1)
vazao_teste = np.array([5,2,7,30,50]).reshape(-1,1)
nivel_previsto = modelo.predict(vazao_teste)
```

```python
import matplotlib.pyplot as plt
```

```python
# Imprimir as previsões

for i in range(len(vazao_teste)):
    print(f'Vazão: {vazao_teste[i][0]} m3/min - Nivel privisto: {nivel_previsto[i]} metros')

plt.scatter(vazao, nivel, color='blue', label='Dados de treinamentos') # gráficos com pontos
plt.plot(vazao_teste, nivel_previsto, color='red', label='Dados de teste')
plt.xlabel('Vazão de água (3m/min)')
plt.ylabel('Nível do tanque (metros)')
plt.title('Relação entre vazão de água e nível do tanque')
plt.legend() # exibe a legenda do gráfico
plt.grid(True) # exibe o grid do gráfico
plt.show()
```
